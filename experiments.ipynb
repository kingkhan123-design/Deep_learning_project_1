```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#Encode Categorical variables
label_encoder_gender=LabelEncoder()
df['Gender']=label_encoder_gender.fit_transform(df['Gender'])
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo=OneHotEncoder()
```

```python
geo_encoder=onehot_encoder_geo.fit_transform(df[['Geography']]).toarray()
print(geo_encoder)
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
#divide the data in dependent and Independent
x=df.drop(['Exited'],axis=1)
y=df['Exited']
```

```python
x
```

```python
## Split the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import  StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
df.head()
```

### ANN Implementation


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
## Build ANN Model
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)), ## First HL and Connected to input laayer
    Dense(32,activation='relu'), ##HL 2
    Dense(1,activation='sigmoid')## Output layer
]
)
```

```python
model.summary()
```

```python
## Compile thhee model
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()

```

```python
model.compile(optimizer=opt,loss="binary_crossentropy",metrics=['accuracy'])
```

```python
## Setup the tensorboard
log_dir='logs/fit'+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
##setup for early stopping
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=20,restore_best_weights=True)
```

```python
### Train the model
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=50,
    callbacks=[tensorflow_callback,early_stopping_callback]
)
```

```python
model.save('model.h5')
```
